Dataframe looks in ok shape, except for the Arson column. Seems like its missing values, so I'll add the mean to entries without a value

```python
crime['Arson3'].dropna().mean()
```

```python
crime['Arson3'] = crime['Arson3'].fillna(crime['Arson3'].dropna().mean())
```

```python
crime = crime.rename(columns={'Murder_and_nonnegligent_manslaughter': 'Murder', 'Arson3': 'Arson'})
```

```python
crime.head()
```

```python
crime['Population'] = crime['Population'].str.replace(',', '')
```

```python
crime['Population'] = crime['Population'].astype(int)
```

```python
crime['Murder'] = crime['Murder'].astype(int)
```

```python
crime['Robbery'] = crime['Robbery'].str.replace(',', '')
crime['Robbery'] = crime['Robbery'].astype(int)
```

```python
crime.head()
```

```python
crime['Murder_feature'] = crime['Murder'].apply(lambda x: 1 if x>0 else 0)
```

```python
crime['Robbery_feature'] = crime['Robbery'].apply(lambda x: 1 if x>0 else 0)
```

```python
crime['Population_feature'] = crime['Population'].apply(lambda x: x**2)
```
