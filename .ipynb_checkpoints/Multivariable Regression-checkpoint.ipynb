```python
%matplotlib inline
import numpy as np
import pandas as pd
import scipy
import sklearn
import matplotlib.pyplot as plt
import seaborn as sns
import math
from sklearn import linear_model
import statsmodels.formula.api as smf
```

```python
crime = pd.read_csv('NEW_YORK-Offenses_Known_to_Law_Enforcement_by_City_2013 - 13tbl8ny.csv')
```

```python
crime.head()
```

```python
crime.drop([0,1,2,3],axis=0,inplace=True)
```

```python
crime.columns=crime.loc[3]
```

```python
crime.tail()
```

```python
crime.drop([352,353,354],axis=0,inplace=True)
```

```python
crime.info()
```

```python
crime.columns
```

```python
crime.drop('Rape\n(revised\ndefinition)1',axis=1, inplace=True)
```

```python
crime = crime.rename(columns={'Violent\ncrime': 'Violent_crime', 'Murder and\nnonnegligent\nmanslaughter': 'Murder_and_nonnegligent_manslaughter','Rape\n(legacy\ndefinition)2': 'Rape', 'Aggravated\nassault': 'Aggravated_assault','Property\ncrime': 'Property_crime', 'Larceny-\ntheft': 'Larceny_theft', 'Motor\nvehicle\ntheft': 'Motor_vehicle_theft'})
```

```python
crime.head()
```

```python
crime.reset_index(drop=True,inplace=True)
```

```python
crime.head()
```

```python
crime.columns
```

```python
crime.info()
```

Dataframe looks in ok shape, except for the Arson column. Seems like its missing values, so I'll add the mean to entries without a value

```python
crime['Arson3'].dropna().mean()
```

```python
crime['Arson3'] = crime['Arson3'].fillna(crime['Arson3'].dropna().mean())
```

```python
crime.info()
```

```python
crime.head()
```

```python
crime.columns
```

```python
crime = crime.rename(columns={'Murder_and_nonnegligent_manslaughter': 'Murder', 'Arson3': 'Arson'})
```

```python
crime.head()
```

```python
crime['Population'] = crime['Population'].str.replace(',', '')
```

```python
crime['Population'] = crime['Population'].astype(int)
```

```python
crime['Murder'] = crime['Murder'].astype(int)
```

```python
crime['Robbery'] = crime['Robbery'].str.replace(',', '')
crime['Robbery'] = crime['Robbery'].astype(int)
```

```python
crime.head()
```

```python
crime['Murder_feature'] = crime['Murder'].apply(lambda x: 1 if x>0 else 0)
```

```python
crime['Robbery_feature'] = crime['Robbery'].apply(lambda x: 1 if x>0 else 0)
```

```python
crime['Population_feature'] = crime['Population'].apply(lambda x: x**2)
```

```python
crime.head()
```

```python
crime_df = crime[['Property_crime','Population','Murder_feature','Robbery_feature','Population_feature']]
```

```python
crime_df.head()
```

```python
regr = linear_model.LinearRegression()
```

```python
X = crime_df[['Population','Murder_feature','Robbery_feature','Population_feature']]
Y = crime_df['Property_crime']
```

```python
regr.fit(X,Y)
```

```python
print('\nCoefficients: \n', regr.coef_)
print('\nIntercept: \n', regr.intercept_)
print('\nR-squared:')
print(regr.score(X, Y))
```

```python
predicted = regr.predict(X).ravel()
actual = crime_df['Property_crime']

residual = actual - predicted

plt.hist(residual)
plt.title('Residual counts')
plt.xlabel('Residual')
plt.ylabel('Count')
plt.show()
```

Seems like the model is overfitted, with a very strong influence by the murder feature. Most of the points aren't errors, about 300/380 are perfect predictions, so its not really normally distributed

```python
plt.scatter(predicted, residual)
plt.xlabel('Predicted')
plt.ylabel('Residual')
plt.axhline(y=0)
plt.title('Residual vs. Predicted')
plt.show()
```
